```python
cat_col=df.select_dtypes(include=['object']).columns.to_list()
num_col=df.select_dtypes(exclude=['object']).columns.to_list()
num_col,cat_col
```

```python
for col in cat_col:
    plt.figure(figsize=(16,8))
    sns.countplot(data=df,x=df[col],color='purple')
    plt.xticks(rotation=45)
    plt.show()
```

```python
num_col.remove("Exceeded_Recommended_Limit")
```

```python
Q1=df["Avg_Daily_Screen_Time_hr"].quantile(0.25)
Q3=df["Avg_Daily_Screen_Time_hr"].quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
df=df[(df["Avg_Daily_Screen_Time_hr"]<=upper_bound) & (df["Avg_Daily_Screen_Time_hr"]>=lower_bound)]
df["Avg_Daily_Screen_Time_hr"].describe()
```

```python
for col in num_col:
    plt.figure(figsize=(12,4))
    sns.boxplot(data=df,y=df[col],color='green')
    plt.show()
```

```python
#  Grouped Bar Chart - Urban vs Rural Screen Time
plt.figure(figsize=(8,6))
sns.barplot(data=df, x="Urban_or_Rural", y="Avg_Daily_Screen_Time_hr", hue="Gender")
plt.title("Average Screen Time by Urban/Rural & Gender")
plt.show()
```

```python
#  Stacked Bar - Device usage by Urban/Rural
device_counts = pd.crosstab(df["Primary_Device"], df["Urban_or_Rural"])
device_counts.plot(kind="bar", stacked=True, figsize=(8,6))
plt.title("Primary Device Usage by Urban/Rural")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()
```
